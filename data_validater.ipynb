```python
# use this to change 2244 to 2324 and add new tabless to db
conn = sqlite3.connect('data/temp/postcovid-current-season.sqlite')
for table in table_list:
    query = f'SELECT * from {table}'
    df = pd.read_sql(query, con=conn)
    
    gids = [str(gid) for gid in df['game_id']]
    gids = [gid.replace('2224', '2324') for gid in gids]
    df['game_id'] = gids

    query = f'DELETE from {table}'
    cursor = conn.cursor()
    cursor.execute(query)
    conn.commit()

    df.to_sql(f'{table}', con=conn, if_exists='append', index=False)
```

```python
for csv in csv_list:
    conn = sqlite3.connect(csv)
    for table in fix_tables:
        query = f'SELECT * from {table}'
        df = pd.read_sql(query, con=conn)
        # # account for team name change following the 2013-2014 season
        # # game_info['away_team'].replace('CHA', 'CHO', inplace=True)
        # # game_info['home_team'].replace('CHA', 'CHO', inplace=True)
        gids = [int.from_bytes(x, byteorder='little') for x in df['game_id']]
        df['game_id'] = gids

        query = f'DELETE from {table}'
        cursor = conn.cursor()
        cursor.execute(query)
        conn.commit()
        
        df.to_sql(f'{table}', con=conn, if_exists='append', index=False)


    
    conn.close()
            
```
